## 1. ILAS

### Summary
ILAS builds Vmin detail and summary outputs from AQUA stacked `TEST_NAME` and `TEST_RESULT` data, with strict parsing and mapping rules. It focuses on correct VMIN extraction, `VMINFWCFG` positional mapping, and proper FMIN separation so aggregated F-corner results remain accurate.

### What It Produces
- ILAS Vmin detail CSV outputs
- ILAS Vmin summary CSV outputs
- Clean separation between VMIN and FMIN analysis paths

### Methodology (High Level)
1. Pull stacked ILAS data from AQUA sources.
2. Parse and normalize Vmin-related test prints.
3. Apply positional `VMINFWCFG` mapping.
4. Exclude FMIN from the main F-corner aggregation.
5. Emit detail and summary artifacts for downstream use.

### Trigger Mode
- Usually automation-triggered as part of weekly workflows
- Can also be run manually when needed

## 2. Weekly Automation

### Summary
Weekly automation runs the NVLH Classhot UPSVF flow, conditionally triggers ILAS analysis, and merges ILAS enrichment columns directly into the final weekly CSV. It is designed for reliability, isolation of failures, and frequency-scoped cleanup.

### What It Produces
- Weekly cleaned/exported UPSVF data
- ILAS-enriched final weekly CSV outputs
- Run logs and reproducible automation behavior

### Methodology (High Level)
1. Pull weekly AQUA data for the configured scope.
2. Clean/filter units and produce base report artifacts.
3. Conditionally run ILAS analysis.
4. Merge ILAS columns into final weekly output.
5. Perform cleanup and write run/report outputs.

### Trigger Mode
- Scheduled (primary mode)
- Can be run manually out of cycle

## 3. Workspace Organizer

### Summary
Workspace organizer keeps scripts maintainable by grouping files into goal-based topic folders, bundling related artifacts together, archiving temporary items, and preserving authored or cloned folders as intended.

### What It Produces
- Organized topic-based script structure
- Archived temporary artifacts instead of deletion
- Workspace organization report

### Methodology (High Level)
1. Initialize topic taxonomy folders and subfolders.
2. Preserve authored or cloned folders as protected.
3. Route known scripts into relevant topic folders.
4. Archive tmp or temporary items.
5. Export organization report and status.

### Trigger Mode
- Manual and out-of-cycle friendly
- Runnable via central control hub
