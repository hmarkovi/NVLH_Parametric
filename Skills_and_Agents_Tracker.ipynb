## Golden Release v1.0 - Parallel Architecture with Daily 5am Automation

**Status**: ✅ GOLDEN - Production Ready (2026-06-23)

### Release Summary
This version consolidates the NVLH parametric analysis pipeline into a reliable, parallelized architecture with:
- **Parallel UPSVF + ILAS execution** with stage manifest for resumption capability
- **Daily 5am automated scheduling** via Windows Task Scheduler
- **Clean output naming** using YYYYMMDD date format (not WW-based)
- **Smart file cleanup** - keeps only final merged CSV, removes intermediate artifacts
- **1-day discovery lookback** for daily automated runs
- **End-to-end validation** with detailed step logging

### Key Components

#### 1. **Orchestrator** (`run_upsvf_ilas_parallel.ps1`)
- Coordinates parallel UPSVF and ILAS pulls
- Maintains stage manifest for resumption without re-pulling
- Skips discovery (uses hardcoded VpoList for daily automation)
- Timeout protection: 30-minute ILAS job limit
- Triggers merge only if ILAS succeeds

**Key Features**:
- Stage manifest tracks: discovery, upsvf_pull, ilas_pull, merge
- Each stage can be individually restarted via `ForceRestartStage` parameter
- Status: IN_PROGRESS, SUCCESS, FAILED, WAITING_FOR_DATA, SKIPPED

#### 2. **UPSVF Script** (`aqua_nvlh_weekly_pull.ps1`)
- Pulls UPSVF data from AQUA with Program Name=NVLHM* filter
- Cleans and filters to DevRevStep_CLASSHOT non-empty units
- **NEW**: Output naming uses date format `Vmin_YYYYMMDD.csv`
- **NEW**: Automatic cleanup of intermediate files (_raw_, _clean_, _merged_work_)
- Preserves 100% of rows through filtering pipeline

#### 3. **ILAS Script** (`aqua_nvlh_ilas_vmin_analysis.ps1`)
- Parses stacked ILAS VMIN_DTS format from AQUA
- Builds detail and summary CSVs per VisualID+LotFromFs
- AquaExe path caching for unattended execution
- Filters by OPERGROUP=6248_CLASSHOT, removes _it and _scrb suffix rows

#### 4. **Merge Script** (`merge-ilas-into-upsvf.ps1`)
- Merges ILAS summary columns into UPSVF clean CSV
- Preserves 100% of UPSVF rows (matched ILAS rows get enriched, unmatched get empty ILAS columns)
- Output: `Vmin_YYYYMMDD_merged.csv`

#### 5. **Validation Script** (`validate_parallel_architecture.ps1`)
- End-to-end validation with 5 checks
- Generates structured JSON report + timestamped log
- Saves all outputs to R:\Products\NVL\NVL-H\Weekly Runs\validation\[VALIDATION_ID]\

#### 6. **Daily Scheduler Setup** (`schedule-daily-5am-automation.ps1`)
- Creates Windows Task Scheduler task: **NVLH-Parametric-Daily-5am**
- Runs every day at **05:00 AM** (user configurable)
- Executes as SYSTEM with administrator privileges
- Automatically restarts on failure with exponential backoff

### Configuration

#### Daily Automation Run
```powershell
# Manually trigger daily run
Start-ScheduledTask -TaskName "NVLH-Parametric-Daily-5am"

# Check status
Get-ScheduledTask -TaskName "NVLH-Parametric-Daily-5am" | Select-Object State, LastRunTime, NextRunTime
```

#### Parameters
- **OutputDirectory**: R:\Products\NVL\NVL-H\Weekly Runs
- **VpoList**: 6248_CLASSHOT (hardcoded for daily runs, no discovery)
- **LastNDaysTestEnd**: 1 (pulls past 24 hours of data)
- **Schedule**: 05:00:00 every day (configurable in setup script)

### Support & Troubleshooting

**Task failed to run at 5am?**
- Check Task Scheduler history: `Get-ScheduledTaskInfo -TaskName "NVLH-Parametric-Daily-5am"`
- Verify account permissions: Task runs as SYSTEM with admin privileges
- Check output directory accessibility: Ensure R:\ drive is mounted and writable

**Merge stage failed but UPSVF succeeded?**
- Use resumption: `run_upsvf_ilas_parallel.ps1 ... -ForceRestartStage merge`
- ILAS may have timed out; increase timeout or check AQUA availability

**Row count mismatch?**
- Check validation report: `R:\Products\NVL\NVL-H\Weekly Runs\validation\[latest]\validation_report.json`
- All UPSVF rows should appear in merged CSV with unmatched rows having empty ILAS columns
