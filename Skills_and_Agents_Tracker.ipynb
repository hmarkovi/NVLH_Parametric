## 1. ILAS

### Summary
ILAS builds Vmin detail and summary CSVs from AQUA stacked `TEST_NAME` and `TEST_RESULT` data, with strict VMIN parsing, VMINFWCFG positional mapping, FMIN separation, and exclusion of FMIN from the main F-corner aggregation.

### What It Produces
- ILAS Vmin detail CSV outputs
- ILAS Vmin summary CSV outputs
- Clean separation between VMIN and FMIN analysis paths

### Methodology (High Level)
1. Pull stacked ILAS data from AQUA sources.
2. Parse and normalize Vmin-related test prints.
3. Apply positional `VMINFWCFG` mapping.
4. Exclude FMIN from the main F-corner aggregation.
5. Emit detail and summary artifacts for downstream use.

### Trigger Mode
- Usually automation-triggered as part of weekly workflows

## 2. Weekly Automation

### Summary
Weekly automation runs the NVLH Classhot UPSVF flow, conditionally triggers ILAS analysis, and merges ILAS enrichment columns directly into the final weekly CSV. It is designed for reliability, failure isolation, and frequency-scoped ILAS cleanup.

### What It Produces
- Weekly cleaned/exported UPSVF data
- ILAS-enriched final weekly CSV outputs
- Run logs and reproducible automation behavior

### Methodology (High Level)
1. Pull weekly AQUA data for the configured scope.
2. Clean and filter units, then produce base report artifacts.
3. Conditionally run ILAS analysis after valid UPSVF output exists.
4. Merge ILAS columns into the final weekly output.
5. Perform cleanup and write run/report outputs.

### Trigger Mode
- Scheduled (primary mode)
- Can be run manually out of cycle
